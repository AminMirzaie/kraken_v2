first we should choose the root of the tree for example bacteria. I found every leaves of this tree and write the taxids of that node in the taxids.txt file in this directory. this all done by code that I wrote in downloadGenomes.py


```python
%%bash
wget "ftp://ftp.ncbi.nlm.nih.gov/genomes/genbank/bacteria/assembly_summary.txt" -P "../resources/"
```

```python
%%bash
python downlaodGenomes.py "Acidobacteria"
```

now we should download genomes of all the organisms that theire taxids is in taxids.txt
this task done by package ncbi-genomes-download
this command has option : **complete genomes**.
this option give me the genomes that is completely sequenced

```python
%%bash
ncbi-genome-download --format fasta --taxid taxids.txt --assembly-level complete bacteria -o ../genomes/
```

now we have all the genomes in genomes folder categorized in refseq folder
i removed all the Extra path and unzip and move all the fasta file to genomes folder by these commands:

```python
%%bash
gunzip ../genomes/refseq/bacteria/*/*.gz
mv ../genomes/refseq/bacteria/*/*.fna ../genomes/
rm -r ../genomes/refseq/
```
